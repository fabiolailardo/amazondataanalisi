```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Histogram o Istogramma kind="hist"
Uso tipico: un istogramma Ã¨ un grafico statistico che rappresenta la distribuzione di frequenze di una variabile continua, dividendo i dati in intervalli detti "bin" e mostrando con barre adiacenti quante osservazioni cadono in ciascun intervallo.â
L'altezza delle barre indica la frequenza assoluta (numero di valori) o relativa (proporzione), mentre la base copre un range di valori sull'asse x; l'area totale Ã¨ proporzionale al totale delle osservazioni.
- <b>hist</b>: istogramma di una o piÃ¹ colonne numeriche per visualizzare la distribuzione (frequenze per intervalli di valori).



### Differenze chiave
Rispetto al <b>grafico a barre</b>, l'istogramma ha barre contigue ovvero senza spazi, dimostra la continuitÃ  dei dati, e serve a esplorare la <b>forma della distribuzione</b>:
- distribuzione simmetrica (graficamente assomiglia ad una campana)
- distribuzione asimmetrica con outlier e long tail/coda lunga

Dimostra:
- Centro: modalitÃ  o picco piÃ¹ alto
- Dispersione: larghezza complessiva (rappresentata nell'asse x)
- Forma: simmetrica, code lunghe o buchi


#### Centro (moda / picco)
Il centro Ã¨ la zona dove le barre sono piÃ¹ alte: Ã¨ lâintervallo di valori in cui si concentrano piÃ¹ osservazioni.

In pratica:
- Se il picco sta intorno a 20, vuol dire che la maggior parte dei dati Ã¨ vicino a 20.
- DISTRIBUZIONE BIMODALE: Se ci sono due picchi (due zone alte separate), la distribuzione Ã¨ bimodale: i dati si raggruppano in due regioni distinte.

```python
np.random.seed(42) #seed -> quante volte rende ripetibile il numero casuale
# Dati di "altezza" in cm, centrati su 170
altezza = pd.Series(np.random.normal(loc=170, scale=5, size=500))

altezza
```

```python
altezza.plot(
    kind="hist",
    bins=20,
)
plt.show()
```

Cosa osservare:
- Il picco delle barre (moda visiva) sarÃ  intorno a 170 cm: quello Ã¨ il centro della distribuzione.
- Media e mediana saranno molto vicine: tipico di una distribuzione simmetrica âa campanaâ.

#### Dispersione: istogramma stretto vs largo
La dispersione misura quanto sono sparsi i dati rispetto al centro.

```python
# Due gruppi con stesso centro (170) ma diversa dispersione
basso_scat = pd.Series(np.random.normal(loc=170, scale=2, size=500))
alto_scat = pd.Series(np.random.normal(loc=170, scale=20, size=500))
display(basso_scat)
display(alto_scat)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,2)
basso_scat.plot(
    kind='hist',
    bins=30,
)
plt.subplot(1,2,1)
alto_scat.plot(
    kind='hist',
    bins=30,
)
plt.tight_layout() #per non accavallare i grafici
plt.show()
```

Cosa osservare:
- Nel primo istogramma le barre si estendono molto di piÃ¹ sull'asse x quindi Ã¨ presente una <b>alta dispersione</b> ovvero altra variabilitÃ 
- Nel secondo istogramma le barre sono piÃ¹ concentrate quindi Ã¨ presente una <b>bassa dispersione</b> ovvero sono presenti tanti dati/valori simili tra loro

#### Forma (simmetria, code, buchi)
La forma descrive il profilo generale dell'istogramma.
- Simmetria: se la parte sinistra e la parte destra del picco del picco sono simili, la distribuzione Ã¨ simmetrica.
- Asimmetrica: se o la parte destra (right skewed) o la parte sinistra (left skewed) hanno delle barre che vanno molto lontano rispetto all'altra.
- Code lunghe: indica che pochi valori molto lontani dal centro e potrebbero essere potenziali outlier. Ad esempio redditi molto alti rispetto alla media del reddito.
- Buchi (gaps): se ci sono degli intervalli senza barre, significa che in questi range non ci sono osservazioni, spesso indicano che i dati si organizzano in gruppi separati. Ad esempio gli studenti in 2 fasce d'etÃ  distinte.

##### Distribuzione Simmetrica
Osservazione: Picco centrale, code uguali a sinistra/destra. Media â mediana.â

```python
np.random.seed(42)
altezza=pd.Series(np.random.normal(170,8,1000))
altezza.plot(
    kind='hist',
    bins=30
)
plt.show()
```

Media e Mediana

```python
print("Media: ", round(altezza.mean(),1))
print("Mediana:", round(altezza.median(),1))
```

##### Asimmetria con coda a destra (right skewed)
Osservazione: Coda lunga a destra, media > mediana (tirata dai valori alti).

```python
reddito = pd.Series(np.random.lognormal(10,1,1000))

reddito.plot(
    kind='hist',
    bins=40,
)
plt.show()
```

Media e mediana

```python
print("Media: ", round(reddito.mean(),1))
print("Mediana:", round(reddito.median(),1))
```

##### Asimmetria con coda a sinistra (left skewed)
Osservazione: Coda lunga a sinistra, media < mediana

```python
np.random.seed(42)
normale = np.random.normal(80,2,1000)
normale2 = np.random.normal(0,9,1000)
tempi = pd.Series(np.clip(100 - normale + normale2, 0, 100))
tempi.plot(
    kind='hist',
    bins=20,
)
plt.show()
```

Media e Mediana

```python
print("Media: ", round(tempi.mean(),1))
print("Mediana:", round(tempi.median(),1))
```

#### Buchi (distribuzione bimodale)

## Boxplot 
Uso Tipico: serve a visualizzare la distribuzione di una variabile numerica attraverso quartili, mediana e outlier e permette dei confrotti rapidi tra gruppi o categorie.
- la scatola centrale che rappresenta l'intervallo interquartile IQR
    - dal 1Â° quartile Q1,25%
    - fino al 3Â° quartile Q3,75%
    - nel centro Q2 valore centrale rappresentato con linea spessa 
- baffi si estendono fino ai valori minimi e massimi entro 1.5 x IQR delle estremitÃ  della scatola, in questa soglia appaioni anche i punti outlier come cerchi.

<b>box</b>: boxplot per riassumere distribuzione (mediana, quartili, outlier) di una o piÃ¹ colonne numeriche

```python
np.random.seed(42)
dff = pd.DataFrame({
    'reparto':np.repeat(['Vendite','IT','HR'],[30,30,30]),
    'stipendio':np.concatenate([np.random.normal(35000,5000,30),np.random.normal(50000,10000,30),np.random.normal(40000,4000,30)])
})
dff
```

### Sintatti personalizzata
- color: sia delle scatole che dei baffi
- sym: simbolo dell'outlier
    - inizialedelcoloresimbolo
- notch: incavo della mediana
- vert: orientamento False=orizzontale True=verticalie
- figsize
- fontsize
- grid: True o False
- patch_artist: True e False per riempire le scatole

### Metodo boxplot
Parametri principali:
- column: colonne numeriche da plottare (di default in programma plotterÃ  tutte le colonne numeriche)
- by: colonna categoria per raggruppare i dati (simile al group by)

```python
dff.boxplot(
    column='stipendio',
    by='reparto',
    sym='rx',
    vert=True,
    notch=True,
    fontsize=10,
    patch_artist=True,
    color='grey',
    figsize=(10,5)
)
ax.tick_params(axis='x', labelbottom=True, labelsize=12, pad=10)
ax.tick_params(axis='y', labelleft=True, labelsize=12, pad=10)
#comando per crea grafico e salvarlo
plt.savefig('grafico.png', dpi=300, bbox_inches='tight')
plt.tight_layout() #per non accavallare i grafici
plt.show()
```

## Scatter plot
Visualizza la relazione tra 2 variabili numeriche (x e y) con punti, rileva:
- correlazioni
- cluster
- trend

<b>scatter</b>: rappresenta la relazione tra due variabili numeriche, con opzioni per usare una colonna come colore (c) e una per la dimensione (s).

### Sintatti Base
Usa df.plot.scatter(x,y) o df.plot(kind='scatter',x,y)

Parametri chiave:
- x: colonna numerica per gli assi
- y: colonna numerica per gli assi
- c: colonna/array per colore
- s: colonna/scalare per dimensione punti
- alpha: trasparenza (0-1)
- figsize: dimensione della figura
- cmap: set di colori per la colonna colori

### HEXBIN plot
crea una mppa a nido d'ape mostra la densitÃ  di punti un piano 2d, divide lo spazio in esagoni e li colora per numero di punti.
Nel momento che abbiamo grandi dataset/dataFrame questo Ã¨ piÃ¹ utile del scatter plot

### Sintatti base
Usa la sintatti:
- df.plot.hexbin(x,y)
- df.plot(kind='hexbin',x,y)

Parametri principali:
- x, y: <b>colonne numeriche</b> per assi
- gridsize: numero degli esagoni (di default 100) piÃ¹ alto Ã¨ il numero piÃ¹ dettagli vedo
- cmap: scala di colori per densitÃ  esempio: 'hot' o 'viridis'
-mincnt: esagoni con < N  punti non mostrati

```python
#DataFrame con posizione dei clienti espressa in lat/log delle cittÃ 
dfcitta = pd.DataFrame({
    'lat':np.random.normal(45.5, 0.5, 5000),
    'log': np.random.normal(9.2, 0.5, 5000),
    'visite': np.random.randint(1,100,5000)
})
dfcitta
```

```python
ax = dfcitta.plot.hexbin(
    x='log',
    y='lat',
    gridsize=30,
    cmap='hot',
    mincnt=10,
    figsize=(10,4)
)
plt.show()
```

```python
fig, (ax1, ax2)= plt.subplots(1, 2, figsize=(15,6) )
dfcitta.plot.hexbin(
    x='log',
    y='lat',
    gridsize=30,
    cmap='hot',
    mincnt=10,
    ax=ax1
)
dfcitta.plot.scatter(
    x='log',
    y='lat',
    c='visite',  #MAPPA I COLORI IN BASE ALLE Viste
    s=10,
    cmap='coolwarm',
    marker='s',
    linewidth=1.5,
    edgecolor='black',
    ax=ax2
)
plt.show()
```
